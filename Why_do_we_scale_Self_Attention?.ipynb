# Generate the softmax function:



```python
def softmax(x):
  # Ensure numerical stability by substracting the max value fromn the each element:
  e_x = np.exp(x- np.max(x))
  softmax_values = e_x / e_x.sum(axis=0) #Normalize
  percent_values = softmax_values *100 #Convert to percentage

  # Format each value to a string with  the decimal 2 places followed by a percentage sign
  formatted_percent_values = [f"{value:.2f}%" for value in percent_values]
  return formatted_percent_values
```
