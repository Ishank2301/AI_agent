```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,LSTM,GRU, Bidirectional
```

```python
# Load the IMDB dataset:
(x_train,y_train), (x_test,y_test) = imdb.load_data(num_words=10000)

# Padding sequences to have the same length:
maxlen = 100 # Maximum sequence length:
x_train = pad_sequences(x_train, maxlen=maxlen, padding='post', truncating='post')
x_test = pad_sequences(x_test, maxlen=maxlen, padding='post', truncating='post')
```

```python
# Build the bidirectional RNN model:
embedding_dim = 32 # Dimension of the emfedding layer:

num_words = 10000 # Define num_words based on the imdb.load_data call

# Define the RNN model:
model = Sequential([
    Embedding(input_dim=num_words,output_dim=embedding_dim),
    SimpleRNN(5),
Dense(1,activation='sigmoid')])


model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train, validation_split=0.2,epochs=5,batch_size=32)
```

# Creating a Bidirectional Model:


```python
model = Sequential([
    Embedding(input_dim=num_words,output_dim=embedding_dim),
    Bidirectional(SimpleRNN(5)),
Dense(1,activation='sigmoid')])

model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

model.summary()
```

Creating a Biredirectional GRU model:

```python
model = Sequential([
    Embedding(input_dim=num_words,output_dim=embedding_dim),
    Bidirectional(GRU(5)),
Dense(1,activation='sigmoid')])

model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])

model.summary()
```

**We actually don't use bidirectional LSTM's as they are already a very complex model adding forward and backward state makes it more complicated**
