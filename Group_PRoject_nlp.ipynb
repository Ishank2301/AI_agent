```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cleaned_resumes.csv')
df.head()
```

#### Checking the column

```python
df['resume_text']
```

#### LowerCaseing

```python
# text normalization
df['resume_text_clean'] = df['resume_text'].str.lower()
df['resume_text_clean']
```

#### Removing stopwords

```python
# load stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
```

```python
# remove stopwords
'''
    |-> split sentence
    |-> remove words in stopwords
    |-> jon back to sentence
'''
def remove_stopwords(text):
    words=text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)
```

```python
df['resume_text_clean'] = df['resume_text_clean'].apply(remove_stopwords)
df['resume_text_clean'].head(2)
```

### TF-IDF

```python
'''
    TF: How often a word appears in a resume

    IDF: Penalizes words that appear in every resume

    Result: important skills get higher weight
'''
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(
    max_features=300
)
```

```python
X_tfidf = tfidf.fit_transform(df['resume_text_clean'])
```

```python
tfidf.get_feature_names_out()[:20]
```

```python
# stopwords are removed but column still have numbers and tfidf learns number as a valid token

# that's why it is giving number in above command
```

```python
# removing numbers
import re

def remove_numbers(text):
    return re.sub(r'\d+', ' ', text)
```

```python
df['resume_text_clean'] = df['resume_text_clean'].apply(remove_numbers)
```

```python
tfidf = TfidfVectorizer(max_features=300)
X_tfidf = tfidf.fit_transform(df['resume_text_clean'])
```
