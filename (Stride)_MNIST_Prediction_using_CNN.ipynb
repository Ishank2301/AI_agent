# Import the  Libraries:

# Using Tensorflow and Keras for Neural Network Programming:

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten
```

```python
(X_train,y_train),(X_test,y_tests) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train[2]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[2])
```

#Creating a CNN Model:

```python
model = Sequential()
# Here the kernel_size determines the filter size the padding same means size of feature map and the image will remain same and not shrink and valid is vice versa .
model.add(Conv2D(32,kernel_size=(3,3), padding='same',strides=(2,1), activation='relu',input_shape=(28,28,1)))
# Here the strides are the shifting of Kernel on the image it signifies the shift of bitwise or matrix row or column wise operation like a (2,1) stride declares a 2 row and 1 column shift.
model.add(Conv2D(32,kernel_size=(3,3), padding='same', strides=(2,1),activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3), padding='same',strides=(2,1),activation='relu'))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

# Model Compilation:

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
from tensorflow .keras.callbacks import EarlyStopping
bepoch = EarlyStopping(monitor='val_accuracy',
                       min_delta=0.00001,
                       patience=5,
                       mode="auto",
                       verbose=1,
                       baseline=None,
                       restore_best_weights=True)
```

```python
history = model.fit(X_train,y_train,epochs=50,validation_split=0.2,callbacks=bepoch)
```

**As we can see due to us increasing the stride the feature maps size was actually reduced even though we were using Padding and we took slightly more epochs to reach the same epochs as we did in Strides=1**

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)   # As we know Sigmoid af output is in probability so to get it in 1,0 we use np.where
from sklearn.metrics import accuracy_score
print(f"Accuracy Score: {accuracy_score(y_tests,y_pred)}")
```

So our predicted output for element from 10 to 20 are these now let's check the real values

```python
import matplotlib.pyplot as plt
import numpy as np

images_to_display = X_test[9:20]
actual_labels = y_tests[9:20]

predicted_labels_raw = model.predict(X_test[9:20])
predicted_labels = predicted_labels_raw.argmax(axis=1)

num_images = images_to_display.shape[0]

rows = int(np.ceil(num_images / 4))
cols = min(num_images, 4)

plt.figure(figsize=(cols * 2, rows * 2))

for i in range(num_images):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(images_to_display[i], cmap='gray')
    plt.title(f"True: {actual_labels[i]}\nPred: {predicted_labels[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
