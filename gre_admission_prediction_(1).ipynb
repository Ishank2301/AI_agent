```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

```

```python
df = pd.read_csv('Admission_Predict.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=['Serial No.'],inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(7,activation='relu',input_dim=7))  # Input dim is the number of features you have like gre score,toefl score and all these.
model.add(Dense(7,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam',metrics=['accuracy'])   # Because We are doing Regression so MSE is loss fxn here.
```

```python
history = model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score,accuracy_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
