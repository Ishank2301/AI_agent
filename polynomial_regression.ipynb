# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df = pd.read_csv('Position_Salaries.csv')
df.info()
df.head()
df.describe()
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```
print(X)
```

```
print(y)
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
#print(predict)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree =4) # First we will try to only have polynoial power of 2 only
X_poly = poly_reg.fit_transform(X) # applying Power of 2 on X.
lin_reg_2 = LinearRegression()  # Applying LinearRegression to X_poly and y to create a polynomial equation(b1x^1,b2x^2).
lin_reg_2.fit(X_poly,y)
```

## Visualising the Linear Regression results

```

plt.scatter(X,y,marker=".",s=200,color = "Red",label="Original Values")
plt.plot(X,lin_reg.predict(X),linewidth=5,color="Purple",label="Prediction")
plt.title("Drawbacks of LinearReg(Over PLR)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

## Visualising the Polynomial Regression results

```

plt.scatter(X,y,marker=".",s=200,color = "Red",label="Original Values")
plt.plot(X,lin_reg_2.predict(X_poly),linewidth=5,color="Purple",label="Prediction")
plt.title("Polynomial Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
