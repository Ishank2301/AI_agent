```python
from sklearn.datasets import make_moons
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import keras
from tensorflow.keras.layers import Dropout, Dense
from tensorflow.keras.optimizers import Adam
import seaborn as sns
from tensorflow.keras.models import Sequential
```

```python
X,y = make_moons(n_samples=300,noise=0.2,random_state=1
                 )
X.shape
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.title('Make Moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
```

# Make a neural network

```python
model = Sequential()
# Input Layer
model.add(Dense(128, input_dim=2,activation='linear',kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.1))
# Hidden Layer
model.add(Dense(128, activation='linear',kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.1))
model.add(Dense(128, activation='linear',kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.1))
model.add(Dense(128, activation='linear',kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.1))
model.add(Dense(128, activation='linear',kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.1))
# output loayer
model.add(Dense(1, activation='sigmoid'))
adam = Adam(learning_rate=0.001)
```

```python
model.summary()
```

```python
# Early stopping to get the best value of epochs:
from tensorflow.keras.callbacks import EarlyStopping
ES = EarlyStopping(monitor = "val_loss",
                   min_delta=0.00001,
                   patience=50,
                   verbose=1,
                   mode="auto",
                   baseline=None,
                   restore_best_weights=True
                   )
```

```python
model.compile(loss="binary_crossentropy",optimizer=adam,metrics=['accuracy','r2_score'])
```

```python
history = model.fit(X,y,validation_split=0.2,epochs=10000,callbacks=ES,verbose=1)
```

```python
history.history
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X, y.ravel(), clf = model, legend=2)
plt.show()
```
