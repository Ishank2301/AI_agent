# Regression Model:

# Only using the linear Activation function to observe the results:

```python
model_1 = Sequential()
model_1.add(Dense(128,activation='linear',input_dim=2,kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model_1.add(Dropout(0.25))
model_1.add(Dense(128,activation='linear',kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model_1.add(Dropout(0.25))
model_1.add(Dense(1,activation='sigmoid'))  # As we know for the output layer in regression activation function is linear
adam = Adam(learning_rate=0.01)
```

```python
model_1.summary()
```

# Compile and Fit the model:

```python
model_1.compile(optimizer=adam,loss="binary_crossentropy",metrics=['mean_squared_error','accuracy'])
```

# Early stopping to Get best Number of epochs:

```python
from tensorflow.keras.callbacks import EarlyStopping
ES = EarlyStopping(
    monitor = "val_loss",
    min_delta=0.00001,
    patience=50,
    verbose=1,
    mode = "auto",
    baseline=None,
    restore_best_weights=True
)
```

```python

history =  model_1.fit(X,y,epochs=500,validation_split=0.2,verbose=1,callbacks=ES)
```

# Evaluate the model:

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X,y.astype('int'),clf=model_1,legend=2)
plt.xlim(-0.7,0.5)
plt.xlim(-0.8,0.8)
plt.show()
```

# Tanh Activation Function:

# Implement Dropout:

*  Create Dropouts for Input and Hidden Layers
*  Dropout based on the size of data



```python
model = Sequential()

model.add(Dense(128, input_dim=2, activation="tanh",kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.5))
model.add(Dense(128, activation="tanh",kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model.add(Dropout(0.5))
model.add(Dense(1, activation="sigmoid"))

adam = Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
model.summary()
```

# Let's try this while using early stopping:

```python
from tensorflow.keras.callbacks import EarlyStopping
ES = EarlyStopping(
    monitor = "val_loss",
    min_delta=0.00001,
    patience=50,
    verbose=1,
    mode = "auto",
    baseline=None,
    restore_best_weights=True
)
```

```python
history = model.fit(X, y, epochs=500, validation_split = 0.2,verbose=1,callbacks=ES)
```

** As we can see after implementing Dropout we only needed to explore till the 7th epoch to get the best result so we can assume that our Dropout model is performing more efficiently than the normal model where every node was in use, so it seems dropping nodes can be helpful in a Neural Network Model.**

```python
plot_decision_regions(X, y.astype('int'), clf=model, legend=2)
plt.xlim(-0.7,0.5)
plt.ylim(-0.8,0.8)
plt.title('p = 0.5')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

**This is the reason we use early stopping saved us computation for nearly 450 epochs as only 93 epochs were required but  we get the best value at the 43rd so you might why not reduce the value of patience, but this has a very significant drawback as doing this might result in a false optimized value of our loss function or the monitor metrics and this is why we should have the patient value between 20-50. **

# ReLU the Best Activation Function:

# Constructing the Neural Network:

```python
model2 = Sequential()
model2.add(Dense(128,activation="relu",input_dim=2,kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model2.add(Dropout(0.25))
model2.add(Dense(128,activation="relu",kernel_regularizer=tf.keras.regularizers.L2(0.03)))
model2.add(Dropout(0.25))
model2.add(Dense(1,activation="sigmoid"))
```
