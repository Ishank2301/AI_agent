# Import Libraries:


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Load the dataset:

```python
df = pd.read_csv('placement.csv')
```

# Observe and Clean the data:

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

# Divide data into dependant and independant features:

```python
X = df.iloc[:,:-1].values
```

```python
print(X)
```

```python
y = df.iloc[:,-1].values
```

```python
print(y)
```

```python
sns.scatterplot(data=df, x='cgpa', y='resume_score', hue='placed')
```

# Implementing Perceptron using sklearn:

```python
from sklearn.linear_model import Perceptron, LogisticRegression

# Initialize Perceptron
P = Perceptron()

# Initialize Logistic Regression
LR = LogisticRegression(random_state=42)
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.pipeline import Pipeline


P.fit(X, y)
```

```python
print("Perceptron Results (from pipeline with scaler):")
# The Perceptron within the pipeline is already fitted by pipe.fit(X,y)
# We can access the fitted Perceptron model directly to get its attributes.
print(f"Perceptron Coefficients: {P.coef_}")
print(f"Perceptron Intercept: {P.intercept_}")

print("\nLogistic Regression Results:")
LR.fit(X,y)
print(f"Logistic Regression Coefficients: {LR.coef_}")
print(f"Logistic Regression Intercept: {LR.intercept_}")
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# Use the classifier from the pipeline for plotting
plot_decision_regions(X, y, clf=P, legend=2)
plt.title('Perceptron Decision Regions')
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay, r2_score
```

```python
# Perceptron Predictions and Confusion Matrix
y_pred_P = P.predict(X)
cm_P = confusion_matrix(y, y_pred_P)

# Logistic Regression Predictions and Confusion Matrix
y_pred_LR = LR.predict(X)
cm_LR = confusion_matrix(y, y_pred_LR)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

disp_P = ConfusionMatrixDisplay(confusion_matrix=cm_P, display_labels=np.unique(y))
disp_P.plot(ax=axes[0])
axes[0].set_title('Perceptron Confusion Matrix')

disp_LR = ConfusionMatrixDisplay(confusion_matrix=cm_LR, display_labels=LR.classes_)
disp_LR.plot(ax=axes[1])
axes[1].set_title('Logistic Regression Confusion Matrix')

plt.tight_layout()
plt.show()
```

```python
accuracy_P = accuracy_score(y, y_pred_P)
print(f"Perceptron Accuracy Score: {accuracy_P}")

accuracy_LR = accuracy_score(y, y_pred_LR)
print(f"Logistic Regression Accuracy Score: {accuracy_LR}")
```
